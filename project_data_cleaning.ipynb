```python
age_median = df['Age'].median()
df['Age'].fillna(age_median, inplace=True)
```

```python
df
```

```python
distance_median = df['DistanceFromHome'].median()
df['DistanceFromHome'].fillna(distance_median, inplace = True)
```

```python
dailyrate_median = df['DailyRate'].median()
df['DailyRate'].fillna(dailyrate_median, inplace = True)
```

```python
enstatisfaction = df['EnvironmentSatisfaction'].median()
df['EnvironmentSatisfaction'].fillna(enstatisfaction, inplace = True)
```

```python
hourlyrate_median = df['HourlyRate'].median()
df['HourlyRate'].fillna(hourlyrate_median, inplace = True)
```

```python
jobinvolvement_median = df['JobInvolvement'].median()
df['JobInvolvement'].fillna(jobinvolvement_median, inplace = True)
```

```python
ncworked_median = df['NumCompaniesWorked'].median()
df['NumCompaniesWorked'].fillna(ncworked_median, inplace = True)
```

```python
totalwyears_median = df['TotalWorkingYears'].median()
df['TotalWorkingYears'].fillna(totalwyears_median, inplace = True)
```

```python
trainingtlyear_median = df['TrainingTimesLastYear'].median()
df['TrainingTimesLastYear'].fillna(trainingtlyear_median, inplace = True)
```

```python
worklifebalance_median = df['WorkLifeBalance'].median()
df['WorkLifeBalance'].fillna(worklifebalance_median, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.columns=['Age', 'Attrition', 'Business Travel', 'Daily Rate', 'Department', 'Distance From Home', 'Education', 'Education Field', 
            'Employee Count',  'Employee Number', 'Environment Satisfaction', 'Gender', 'Hourly Rate', 'Job Involvement', 'Job Level',
            'Job Role', 'Job Satisfaction', 'Marital Status', 'Monthly İncome', 'Monthly Rate', 'Num Companies Worked', 'Over 18',
            'Over Time', 'Percent Salary Hike', 'Performance Rating', 'Relationship Satisfaction', 'Standard Hours', 'Stock Option Level',
            'Total Working Years', 'Training Times Last Year', 'Work Life Balance', 'Years At Company', 'Years In Current Role', 'Years Since Last Promotion',
            'Years With Current Manager']
            
```

```python
df
```

```python
df['Attrition'] = df['Attrition'].replace({
    'y' : 'Yes',
    'n' : 'No' })
```

```python
df
```

```python
df['Business Travel'].unique()
```

```python
df['Business Travel'] = df['Business Travel'].replace({
    'travel_rarely' : 'Rarely', 'travel_frequently' : 'Frequently',
    'nn-Travel' : 'No Travel' })
```

```python
df
```

```python
df['Department'].unique()
```

```python
df['Department'] = df['Department'].replace({
    'sales' : 'Sales', 'research & development' : 'R&D',
    'Human Resources' : 'HR'})
```

```python
df
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].replace({
    'f' : 'Female', 'm' : 'Male'})
```

```python
df['Gender']
```

```python
df
```

```python
df['Marital Status'].unique()
```

```python
df['Marital Status'] = df['Marital Status'].replace({
    'sngl' : 'Single', 'mrrd' : 'Married'})
```

```python
df['Marital Status']
```

```python
df['Over 18'].unique()
```

```python
df['Over 18'] = df['Over 18'].replace({
    'Y' : 'Yes'})
```

```python
df['Over 18']
```

```python
df['Over Time'].unique()
```

```python
df['Over Time'] = df['Over Time'].replace({
    'y' : 'Yes', 'n' : 'No'})
```

```python
df['Over Time']
```

```python
df
```

```python
df['Education'].unique()
```

```python
df['Education'] = df['Education'].replace({
    1 : 'Below Collage', 2 : 'Collage', 3 : 'Bachelor',
     4 : 'Master', 5 : 'Doctor'})
```
